```python
# Linear Regression model
# we have used linear model to analyze anova
#ols=ordinary least square error method
# error decreases,ols=100,ols=50
# d1square+d2square.......dn square
import statsmodels.api as sm
from statsmodels.formula.api import ols
# step1:Import required libraries
# step 2:EDA1(SHAPE,READ,INFO,DESCRIBE,CORELATION AND SO ON)
# STEP3:EDA2(MODEL BUILDING AND MODEL EVALUATION)
```

```python
mod=ols( 'soil_moisture ~ soil_temperature',data=df)
mod
# tidle operator.analysis,data=inbuiltfunction
```

```python
mod=ols( 'soil_moisture ~ soil_temperature',data=df).fit()
mod
```

```python
#.ANOVA_LM
aov=sm.stats.anova_lm(mod)
```

```python
#Multiple LR MODEL
# two way anova(soil temperature and ph-multiple Independent variables,soil moisture is dv)
# add ph and contour and anova
mod=ols( 'soil_moisture ~ soil_temperature+ph',data=df).fit()
mod
```

```python
#.ANOVA_LM
aov1=sm.stats.anova_lm(mod)
```
